```python
import requests

REST_API_KEY = '2fa4da0b370eee3234c565b78ecff325'
# url ì˜ ? ë’¤ìª½ì€ ì¿¼ë¦¬ íŒŒë¼ë¯¸í„°
# keywordì˜ ì£¼ë³€ ë°˜ê²½ 1km ë‚´ì˜ ë§›ì§‘ ì°¾ê¸°
KEYWORD_LOCAL_URL = 'https://dapi.kakao.com/v2/local/search/keyword.json?query={}&radius=1000'

keywords=['ì„œìš¸ì—­', 'í™ëŒ€ì…êµ¬ì—­']
```

```python
# header ì— ì¸ì¦ ì •ë³´ ì¶”ê°€ (ì•ˆí•˜ë©´ 401 error)
headers = {
    'Authorization' : f'KakaoAK {REST_API_KEY}'
}
response = requests.get(KEYWORD_LOCAL_URL.format(keywords[1] + 'ë§›ì§‘'), headers=headers)
response
```

```python
import time
all_data=[]
# ë‚˜ë¨¸ì§€ í˜ì´ì§€ ë° í‚¤ì›Œë“œì— ëŒ€í•œ ë°ì´í„°ë¥¼ ê°€ì ¸ì˜¤ê¸°
for keyword in keywords:
    for page in range(1, 6):  # 1í˜ì´ì§€ë¶€í„° 5í˜ì´ì§€ê¹Œì§€ ë°˜ë³µ
        # ì²« ë²ˆì§¸ í‚¤ì›Œë“œì˜ ì²« í˜ì´ì§€ëŠ” ì´ë¯¸ ìš”ì²­ë˜ì—ˆìœ¼ë¯€ë¡œ ê±´ë„ˆëœë‹ˆë‹¤.
        if keyword == keywords[0] or page == 1:
            continue
        params = {
            'page': page,
            'size': 15  # í˜ì´ì§€ë‹¹ ê²°ê³¼ ìˆ˜
        }
        response = requests.get(KEYWORD_LOCAL_URL.format(keyword + ' ë§›ì§‘'), headers=headers, params=params)
        if response.status_code == 200:
            data = response.json().get('documents', [])
            if not data:  # ë” ì´ìƒ ë°ì´í„°ê°€ ì—†ìœ¼ë©´ ì¤‘ì§€
                break
            all_data.extend(data)  # ë°ì´í„°ë¥¼ ë¦¬ìŠ¤íŠ¸ì— ì¶”ê°€
        else:
            print(f"Error fetching data for {keyword} page {page}: {response.status_code}")
            break
        time.sleep(2)  # ì›¹ í˜ì´ì§€ ë¡œë“œë¥¼ ë³´ì¥í•˜ê¸° ìœ„í•´ 2ì´ˆ ì‰¬ê¸°
# ìˆ˜ì§‘ëœ ëª¨ë“  ë°ì´í„°ë¥¼ ì¶œë ¥
for place in all_data:
    print({
        'address_name': place['address_name'],
        'id': place['id'],
        'place_name': place['place_name'],
    })
```

```python
# ê° ê°€ê²Œì— ë¶€ì—¬ëœ idì™€ ë§ˆì§€ë§‰ comment id ë¥¼ ë„£ì„ ìˆ˜ ìˆê²Œ format ì‚½ì…
COMMENT_URL = "https://place.map.kakao.com/m/commentlist/v/{}/{}?order=USEFUL&onlyPhotoComment=false"
```

```python
# ê°€ê²Œ id ì¶”ì¶œ
ids = [d['id'] for d in all_data]
ids
```

```python
len(ids)
```

```python
all_comments = []
for id in ids:
    # ì²« ë²ˆì¨° ì½”ë©˜íŠ¸ì˜ idëŠ” ë¬´ì¡°ê±´ 0
    comment_id = 0

    # has_nextê°€ true ë©´ ë§ˆì§€ë§‰ ì½”ë©˜íŠ¸ idë¥¼ ìƒˆë¡­ê²Œ ë„£ì–´ ê³„ì† í¬ë¡¤ë§
    has_next = True

    while has_next:
        # ìˆ˜ì§‘í•´ì•¼ í•  url ì„¤ì •
        SCRAP_COMMENT_URL = COMMENT_URL.format(id, comment_id)
        # ë§Œë“¤ì–´ì§„ urlë¡œ ìš”ì²­
        response = requests.get(SCRAP_COMMENT_URL)
        comment_data = response.json()['comment']
        # ëŒ“ê¸€ ë°ì´í„° ê°€ì ¸ì˜¤ê¸°
        comment_list = comment_data['list']
        all_comments.extend(comment_list) # 1ì°¨ì› ë°°ì—´ë¡œ ì­‰ ì´ì–´ë¶™ì´ê¸°
        # ë‹¤ìŒ í˜ì´ì§€ ì¡´ì¬ ì—¬ë¶€ í™•ì¸
        has_next = comment_data['hasNext']
        # has_next ê°€ True ì¸ ê²½ìš° ë§ˆì§€ë§‰ ì½”ë©˜íŠ¸ì˜ idë¥¼ comment_idë¡œ ì„¤ì •
        if has_next:
            comment_id = comment_list[-1]['commentid']

        print(comment_list)

all_comments
```
